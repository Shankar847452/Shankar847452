# 5. Write a program to Visualize the dataset to gain insights using Matplotlib or Seaborn by plotting scatter plots, bar charts.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Admission_Predict.csv')
data
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
# Boxplot for GRE Score
plt.figure(figsize=(8, 6))
sns.boxplot(x='Research', y='GRE Score', data=data)
plt.title('Boxplot of GRE Score by Research Experience')
plt.xlabel('Research Experience')
plt.ylabel('GRE Score')
plt.show()
```

```python
# Scatter plot for TOEFL Score vs. CGPA
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TOEFL Score', y='CGPA', data=data, hue='Research')
plt.title('Scatter plot of TOEFL Score vs. CGPA by Research Experience')
plt.xlabel('TOEFL Score')
plt.ylabel('CGPA')
plt.show()
```

```python
# Bar chart for University Rating
plt.figure(figsize=(8, 6))
sns.countplot(x='University Rating', data=data)
plt.title('Bar chart of University Rating')
plt.xlabel('University Rating')
plt.ylabel('Count')
plt.show()
```
