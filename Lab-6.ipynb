# 6. Write a program to Visualize the dataset to gain insights using Seaborn by plotting Pairplot and Heatmap. Use Admission_Predict.csv file.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('Admission_Predict.csv')
data
```

```python
# Display the first few rows of the dataset
print(data.head())
```

```python
#pairplot
sns.pairplot(data, vars=['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research'],
             hue='Chance of Admit')
plt.suptitle('Pairplot of GRE Score, TOEFL Score, University Rating, SOP, LOR, CGPA, and Research(colored by Admission Chance)')
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of the Dataset')
plt.show()
```
